```python
import networkx as nx
import pickle as pkl
import matplotlib.pyplot as plt 
```

```python
G = nx.DiGraph()
```

```python
with open('../pickle_files/edges.pkl','rb') as f:
    edges = pkl.load(f)
```

```python
G.add_edges_from(edges)
```

```python
assert (len(list(G.nodes)), len(list(G.edges))) == (7115,103689)
```

```python
pr = nx.pagerank(G)
```

```python
pr = {x:y for x,y in list(sorted(pr.items(),key=lambda x:x[0]))}
```

```python
fig, ax1 = plt.subplots(1, 1,figsize=(15,5))
ax1.set_title("Page Rank of all the Nodes")
ax1.bar(pr.keys(),pr.values())
```

```python
Hub, Auth = nx.hits(G)
```

```python
Hub = {x:y for x,y in list(sorted(Hub.items(),key=lambda x:x[0]))}
Auth = {x:y for x,y in list(sorted(Auth.items(),key=lambda x:x[0]))}
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(15,10))
ax1.set_title("Hub Score of all the Nodes")
ax2.set_title("Authority Score of all the Nodes")
ax1.bar(Hub.keys(),Hub.values())
ax2.bar(Auth.keys(),Auth.values())
```
